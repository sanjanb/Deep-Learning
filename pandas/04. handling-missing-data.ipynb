# Converting String to Datetime in Pandas

In real-world datasets, dates are often stored as **strings**.  
Pandas provides the `pd.to_datetime()` function to **convert string columns into datetime objects**, which allows you to perform date-based operations like filtering, sorting, and extracting components (year, month, day).


### 1. Example: Converting a String Column
```python
import pandas as pd

# Sample DataFrame with date as string
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Joining_Date": ["2021-05-14", "2020-08-23", "2022-02-10", "2019-11-30"]
}
df = pd.DataFrame(data)

# Convert 'Joining_Date' to datetime
df["Joining_Date"] = pd.to_datetime(df["Joining_Date"])

print(df)
print("\nData types after conversion:")
print(df.dtypes)

# Handling Missing Data with `.fillna()`

In real-world datasets, some values are often **missing** (represented as `NaN`).  
Pandas provides the **`.fillna()`** method to **replace missing values** with a specific value or a computed value (like mean, median, or method-based filling).


### 1. Basic Example: Fill with a Constant
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [24, np.nan, 22, 32],
    "Salary": [48000, 54000, np.nan, 62000]
}
df = pd.DataFrame(data)

# Fill missing values with a constant
df_filled = df.fillna(0)
print(df_filled)


````python
df1 = df.fillna(0)
df1
````

### Having zero is not best way to handle missing data

````python
df1 = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'No Event'
})

df1
````

# Forward Fill Missing Values with `fillna(method='ffill')`

In time series or sequential datasets, you may want to **fill missing values with the previous valid value**.  
Pandas provides the **forward fill** method (`method='ffill'`) for this purpose.


### 1. Example: Forward Fill
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Date": pd.date_range("2025-10-01", periods=6),
    "Sales": [100, np.nan, 150, np.nan, np.nan, 200]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Forward fill missing values
df_ffill = df.fillna(method="ffill")

print("\nAfter forward fill:")
print(df_ffill)


````python
df1 = df.fillna(method='ffill')
df1
````

# Backward Fill Missing Values with `fillna(method='bfill')`

Backward fill allows you to **replace missing values with the next valid value** in a column.  
This is useful when future data can logically fill in gaps in your dataset.


### 1. Example: Backward Fill
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Date": pd.date_range("2025-10-01", periods=6),
    "Sales": [100, np.nan, 150, np.nan, np.nan, 200]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Backward fill missing values
df_bfill = df.fillna(method="bfill")

print("\nAfter backward fill:")
print(df_bfill)


````python
df1.fillna(method='bfill')
df1
````

# Backward Fill Across Columns (`axis='columns'`)

By default, `fillna(method='bfill')` fills missing values **downward across rows** (`axis=0`).  
Using `axis='columns'` fills missing values **horizontally across columns**, replacing NaNs with the next valid value in the same row.


### 1. Example: Backward Fill Across Columns
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Q1": [100, np.nan, 150, np.nan],
    "Q2": [np.nan, 200, np.nan, 180],
    "Q3": [130, np.nan, 170, np.nan]
}
df1 = pd.DataFrame(data)

print("Original DataFrame:")
print(df1)

# Backward fill across columns
df1_filled = df1.fillna(method="bfill", axis="columns")

print("\nAfter backward fill across columns:")
print(df1_filled)


````python
df1 = df.fillna(method='bfill', axis='columns')
df1
````

# Forward Fill with a Limit (`limit=1`)

The `limit` parameter in `fillna` restricts the **number of consecutive NaNs to fill**.  
Using `method='ffill', limit=1` will **fill only the first missing value** in a sequence, leaving subsequent NaNs untouched.


### 1. Example
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Q1": [100, np.nan, np.nan, 130],
    "Q2": [np.nan, 200, np.nan, 150]
}
df1 = pd.DataFrame(data)

print("Original DataFrame:")
print(df1)

# Forward fill with a limit of 1
df1_ffill_limit = df1.fillna(method="ffill", limit=1)

print("\nAfter forward fill with limit=1:")
print(df1_ffill_limit)


````python
df1 = df.fillna(method='ffill', limit=1)

df1
````

# Interpolating Missing Values with `interpolate()`

`interpolate()` is used to **fill missing (NaN) values using interpolation**, which estimates values based on surrounding data points.  
Itâ€™s especially useful for **time series or numeric datasets** where missing values can be inferred from neighboring values.


### 1. Example: Linear Interpolation
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Day": [1, 2, 3, 4, 5],
    "Sales": [100, np.nan, np.nan, 160, 200]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Interpolate missing values
df_interpolated = df.interpolate()

print("\nAfter interpolation:")
print(df_interpolated)


````python
df1 = df.interpolate()

df1
````

### Notes

* Default method is **linear interpolation**, filling values evenly between known points.
* Can specify different methods:

  ```python
  df.interpolate(method='polynomial', order=2)
  df.interpolate(method='time')                
  ```
* Useful for **time series, sensor readings, or any sequential numeric data**.
* `inplace=True` can modify the original DataFrame directly.

````python
df1 = df.interpolate(method='time')

df1
````

# Dropping Rows with Missing Values Using `dropna()`

The **`dropna()`** method removes rows or columns that contain missing values (`NaN`).  
The `thresh` parameter allows you to **keep rows/columns with at least a minimum number of non-NaN values**.


### 1. Example: Dropping Rows with Few Non-NaNs
```python
import pandas as pd
import numpy as np

# Sample DataFrame with missing values
data = {
    "Q1": [100, np.nan, 150, np.nan],
    "Q2": [np.nan, 200, np.nan, np.nan],
    "Q3": [130, np.nan, 170, np.nan]
}
df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)

# Drop rows with less than 1 non-NaN value
new_df = df.dropna(thresh=1)

print("\nAfter dropna(thresh=1):")
print(new_df)


````python
new_df = df.dropna(thresh=1)
new_df
````
