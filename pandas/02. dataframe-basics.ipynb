# Slicing in Pandas

**Slicing** in Pandas means selecting a specific portion (or slice) of rows and columns from a DataFrame or Series.

It allows you to:
- View only part of your data  
- Focus on specific rows or columns  
- Prepare subsets for further analysis  

Pandas provides several ways to perform slicing:

### 1. Row Slicing (by index)
You can use **index positions** or **labels** to slice rows.

```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Age": [24, 27, 22, 32, 29],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"]
}
df = pd.DataFrame(data)

# Slice rows by index (first 3 rows)
print(df[0:3])

# Conditional Selection in Pandas

Conditional selection allows you to **filter rows** in a DataFrame based on certain conditions.  
This is extremely useful when you want to focus on specific subsets of data.


### 1. Basic Conditional Selection
```python
import pandas as pd

# Sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Age": [24, 27, 22, 32, 29],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"],
    "Salary": [48000, 54000, 50000, 62000, 58000]
}
df = pd.DataFrame(data)

# Select rows where Age > 25
df_age = df[df["Age"] > 25]
print(df_age)

# Setting an Index in Pandas

In Pandas, you can **set one of the DataFrame columns as the index**.  
This helps identify rows with meaningful labels instead of default numeric indices and makes row selection easier.


### 1. Using `set_index()`
```python
import pandas as pd

# Sample DataFrame
data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"]
}
df = pd.DataFrame(data)

# Set 'EmployeeID' as the index
df.set_index("EmployeeID", inplace=True)

# Display the DataFrame
print(df)

# Resetting Index in Pandas

The **`reset_index()`** method in Pandas is used to **reset the row index** of a DataFrame back to the default integer index.  
This is especially useful after setting a custom index or after dropping/adding rows.


### 1. Basic Example
```python
import pandas as pd

# Sample DataFrame with a custom index
data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Department": ["HR", "IT", "Finance", "Marketing", "IT"]
}
df = pd.DataFrame(data)
df.set_index("EmployeeID", inplace=True)

print("DataFrame with custom index:")
print(df)

# Reset index back to default
df_reset = df.reset_index()
print("\nDataFrame after reset_index():")
print(df_reset)
