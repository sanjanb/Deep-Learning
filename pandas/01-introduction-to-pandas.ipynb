# üîß Data Munging and Wrangling

Before we can analyze data, we often need to **clean** and **reshape** it ‚Äî this process is known as **data munging** or **data wrangling**.

These two terms are often used interchangeably and refer to the process of:
- **Collecting** raw data from different sources  
- **Cleaning** it by handling missing values, duplicates, or errors  
- **Transforming** it into a consistent and structured format  
- **Enriching** it by adding or merging useful information from other datasets  

In real-world projects, raw data is rarely ready for analysis. It might have:
- Inconsistent column names or types  
- Missing or duplicate entries  
- Extra spaces, special characters, or incorrect values  

Pandas is especially powerful for data munging and wrangling tasks.  
It provides simple methods to:
- Filter and sort data  
- Fill or drop missing values  
- Merge and join multiple datasets  
- Convert data types  
- Group and aggregate records  

### Example:
```python
import pandas as pd

# Load a dataset
df = pd.read_csv("data.csv")

# Preview the first few rows
df.head()

# Drop missing values
df_clean = df.dropna()

# Convert a column to a consistent type
df_clean["price"] = df_clean["price"].astype(float)

# Check the cleaned data
df_clean.info()
