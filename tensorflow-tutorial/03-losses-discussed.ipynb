```python
y_pred = np.array([1, 1, 0, 0, 1])
y_true = np.array([0.30, 0.7, 1, 0, 0.5])
```

## Mean Absolute Error


```python

def mae(y_true, y_pred):
    total_error = 0
    for yt, yp in zip(y_true, y_pred):
        total_error += abs(yt - yp)
    print(f"Total Error: {total_error}")
    mae = total_error / len(y_pred)
    return mae
```

```python
mae(y_true, y_pred)
```

## Log Loss also known as binary cross entropy or logistic regression

```python
epsilon = 1e-15
```

```python
np.log([0.0000000000000001]) # Should give a nearest value to zero, not exactly zero

# so we define our own term epsilon
np.log([epsilon])
```

```python
y_pred_new = [max(i, epsilon) for i in y_pred] # for 0 value
y_pred_new
```

```python
y_pred_new = [min(i, 1-epsilon) for i in y_pred_new] # for 1 value
y_pred_new
```

```python
y_pred_new = np.array(y_pred_new)
np.log(y_pred_new)
```

```python
-np.mean(y_true*np.log(y_pred_new)+(1-y_true)*np.log(1-y_pred_new))
```
