## Stemming in NLTK

```python
import nltk 
import spacy
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
words = ["eating", "eats", "eat", "ate", "adjustable", "rafting", "ability", "meeting"]

for word in words:
    print(word, "|", stemmer.stem(word))
```

As you can see, the it is usig fixed rules to convert the words to its base word

## Lemmatization in Spacy

```python
nlp = spacy.load('en_core_web_sm')

doc = nlp("Mando talked for 3 hours although talking isn't his thing")
doc = nlp("eating eats eat ate adjustable rafting ability meeting better")
for token in doc:
    print(token, " | ", token.lemma_)
```

```python
nlp = spacy.load('en_core_web_sm')

doc = nlp("Mando talked for 3 hours although talking isn't his thing")
doc = nlp("eating eats eat ate adjustable rafting ability meeting better")
for token in doc:
    print(token, " | ", token.lemma_, "|", token.lemma)
```

As you can see, each have their own hashcode

## Customizing lemmatizer

```python
nlp.pipe_names
```

```python
doc = nlp("Bro you wanna go? Brah, don't say no, I'am exhausted")

for token in doc:
    print(token.text, "|", token.lemma_)
```

All the slangs are taken as the same words only, no changes are taken after the lemmatization, in order to change that, 
- in order to customize this model we use attribute ruler from nlp.pipe_names
- use any custom rule you want
- and wallaaaaa, you will get it<br>
Lets see how this works

```python
ar = nlp.get_pipe('attribute_ruler')

ar.add([[{'TEXT':'Bro'}],[{"TEXT":"Brah"}]],{"LEMMA":"Brother"})
doc = nlp("Bro you wanna go? Brah, don't say no, I'am exhausted")

for token in doc:
    print(token.text, "|", token.lemma_)
```
