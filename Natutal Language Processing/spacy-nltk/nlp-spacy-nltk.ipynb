Load the en core module to do some spcacy oparaions

```python
import spacy

nlp = spacy.load("en_core_web_sm") 
print("spaCy imported successfully!")
```

- Create a object of nlp and use it to do any spacy operations
- Lets convert the document into sentences

```python
doc = nlp("The ridiculously fluffy quantum goat quickly devoured all three philosophical parking meters yesterday. Gave a big poop out of it")

for sentence in doc.sents:
    print(sentence)
```

- It is very easy to convert the sentences into words
- use nested for loop while calling it, just like oops concepts
- these two operations are called,
  - Sentence Tokenization
  - Word Tokenization

```python
for sentence in doc.sents:
    for words in sentence:
        print(words)

# Simple as that
```

## NLTK
it is a sentence tokenizin toool

```python
import nltk
nltk.download('punkt')

print("nltk imported sucessfully")
```

```python
from nltk.tokenize import sent_tokenize

sentences = sent_tokenize("The ridiculously fluffy quantum goat quickly devoured all three philosophical parking meters yesterday. Gave a big poop out of it")
sentences
```

```python
from nltk.tokenize import word_tokenize

words = word_tokenize("The ridiculously fluffy quantum goat quickly devoured all three philosophical parking meters yesterday. Gave a big poop out of it")
words
```

Now you got a difference between Spacy and NLTK
- nltk is a string processing library
- nltk is has to be setup manually to get the better result
- but spacy get the best way to the result
- nltk has to do lot of customization, to get the good results
- spacy is best for devs, but nltk is for researchers
