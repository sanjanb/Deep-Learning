```python
import spacy

nlp = spacy.load('en_core_web_lg')
print("Spacy small model loaded sucessfully")
```

```python
doc = nlp('dog cat banana, kakka')

for token in doc:
    print(token.text, "Vector:", token.has_vector, "OOV:", token.is_oov)
```

```python
base_token = nlp("Bread")
base_token[0].vector.shape
```

```python
# Lets try example with some different context

doc = nlp("bread sandwich burger car tiger human wheat")

for token in doc:
    print(f'{token.text} < - > {base_token.text}', token.similarity(base_token))

# This example gives the meaning of the word with context with the word
```

```python
def print_similarity(base_token, word_to_compare):
    base_token = nlp(base_token)
    doc = nlp(word_to_compare)

    for token in doc:
        print(f'{token.text} < - > {base_token.text}', token.similarity(base_token))
        
```

```python
print_similarity("iphone", "apple samsung iphone nokia dog kitten")
```

```python
king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
women = nlp.vocab["women"].vector
queen = nlp.vocab["queen"].vector

result = king - man + women

result
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([result], [queen])
```
