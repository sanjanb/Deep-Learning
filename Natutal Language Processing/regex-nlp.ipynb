```python
# Age

pattern = 'age (\d+)'

match = re.findall(pattern, text)
match
```

```python
# Name

pattern = 'Born(.*)\n'
match = re.findall(pattern, text)
name = match[0]
name.strip()
```

```python
# DOB

pattern = 'Born.*\n(.*)\(age'
match = re.findall(pattern, text)
name = match[0]
name.strip()
```

```python
# Birth Place

pattern = '\(age.*\n(.*)'

match = re.findall(pattern, text)
name = match[0]
name.strip()
```

```python
def get_pattern_match(pattern, text):
    matches = re.findall(pattern, text)
    if matches:
        return matches[0]
```

```python
get_pattern_match('\(age.*\n(.*)', text)
```

```python
def get_personal_information(text):
    age = get_pattern_match('age (\d+)', text)
    name = get_pattern_match('Born(.*)\n', text)
    dob = get_pattern_match('Born.*\n(.*)\(age', text)
    birth = get_pattern_match('\(age.*\n(.*)', text)

    return {
        'age': int(age),
        'name': name.strip(),
        'birth_date': dob.strip(),
        'birth_place': birth.strip()
    }

get_personal_information(text)
```
